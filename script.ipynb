## Recaudación por Año

#### Total

```python
df = pd.DataFrame(recaudacion_anual)

# Excluir el año 2024 para la preparación de la predicción
df_filtered = df[df['Año'] != 2024]

# Extraer columnas necesarias para la predicción
X = df_filtered[['Año']].values
y_total = df_filtered['Recaudación Total'].values
y_occupancy = df_filtered['Promedio índice de ocupación'].values

# Aumentar el peso para los últimos 3 años
weights = np.ones_like(y_total)
weights[-3:] = 200  # Más peso para los últimos 3 años

# Crear modelos de regresión lineal ponderada
model_total = LinearRegression()
model_occupancy = LinearRegression()

# Ajustar modelos con pesos
model_total.fit(X, y_total, sample_weight=weights)
model_occupancy.fit(X, y_occupancy, sample_weight=weights)

# Predicciones para 2024 y 2025
future_years = np.array([[2024], [2025]])
pred_total = model_total.predict(future_years)
pred_occupancy = model_occupancy.predict(future_years)

# Agregar predicciones al DataFrame para la visualización
df_predictions = pd.DataFrame({
    'Año': [2024, 2025],
    'Recaudación Total': pred_total,
    'Promedio índice de ocupación': pred_occupancy
})

# Combinar datos originales (excluyendo 2024) con predicciones
df_combined = pd.concat([df_filtered, df_predictions], ignore_index=True)

# Mapas de colores personalizados para recaudación total e índice de ocupación
cmap_revenue = mcolors.LinearSegmentedColormap.from_list('revenue_cmap', [(0, 'red'), (0.5, 'orange'), (1, 'green')])
cmap_occupancy = mcolors.LinearSegmentedColormap.from_list('occupancy_cmap', [(0, 'red'), (0.5, 'orange'), (1, 'green')])

# Función para asignar colores basados en la recaudación total
def color_map_revenue(value):
    norm_value = (value - df_combined['Recaudación Total'].min()) / (df_combined['Recaudación Total'].max() - df_combined['Recaudación Total'].min())
    return cmap_revenue(norm_value)

# Función para asignar colores basados en el índice de ocupación
def color_map_occupancy(value):
    norm_value = (value - df_combined['Promedio índice de ocupación'].min()) / (df_combined['Promedio índice de ocupación'].max() - df_combined['Promedio índice de ocupación'].min())
    return cmap_occupancy(norm_value)

# Función formateadora para porcentaje de ocupación
def format_percent(x, _):
    return '{:}%'.format(x)

# Crear gráficos de barras
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

# Gráfico de barras para 'Recaudación Total'
colors_revenue = [color_map_revenue(value) for value in df_combined['Recaudación Total']]
bars1 = ax1.bar(df_combined['Año'], df_combined['Recaudación Total'], color=colors_revenue, edgecolor='black')
ax1.set_title('Recaudación Total por Año')
ax1.axvline(x=2023.5, color='gray', linestyle='--', linewidth=1, label='Frontera de Predicción')
ax1.legend()
ax1.yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f} €'))

# Gráfico de barras para 'Promedio índice de ocupación'
colors_occupancy = [color_map_occupancy(value) for value in df_combined['Promedio índice de ocupación']]
bars2 = ax2.bar(df_combined['Año'], df_combined['Promedio índice de ocupación'], color=colors_occupancy, edgecolor='black')
ax2.set_title('Promedio índice de ocupación por Año')
ax2.axvline(x=2023.5, color='gray', linestyle='--', linewidth=1, label='Frontera de Predicción')
ax2.legend()
ax2.yaxis.set_major_formatter(FuncFormatter(format_percent))

# Mostrar todos los años en el eje x
plt.xticks(df_combined['Año'])
ax2.set_xlabel('Año')

# Ajustar diseño
fig.tight_layout()
plt.show()
```

#### Promedio índice de ocupación por Mes

```python
df_2022 = pd.DataFrame(datos_2022)
df_2023 = pd.DataFrame(datos_2023)
df_2024 = pd.DataFrame(datos_2024)

# Funciones de formato para los ejes y leyendas
def thousands(x, _):
    return '{:,.0f} €'.format(x)

def format_percent(x, _):
    return '{:}%'.format(x)

# Definir meses de invierno y verano
meses_invierno = ['enero', 'febrero', 'noviembre', 'diciembre']
meses_verano = ['junio', 'julio', 'agosto']

# Encontrar índices de los meses de invierno y verano en el DataFrame de 2022
indices_invierno = [df_2022[df_2022['Mes'] == mes].index[0] for mes in meses_invierno]
indices_verano = [df_2022[df_2022['Mes'] == mes].index[0] for mes in meses_verano]

# Crear el gráfico
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 18), sharex=True)

# Gráficos de líneas principales de recaudación total
line1, = ax1.plot(df_2022['Mes'], df_2022['Recaudación Total'], label='2022', marker='o', alpha=0.3)
line2, = ax1.plot(df_2023['Mes'], df_2023['Recaudación Total'], label='2023', marker='o', alpha=0.3, color='brown')
line3, = ax1.plot(df_2024['Mes'], df_2024['Recaudación Total'], label='2024', marker='o', alpha=0.3)
ax1.yaxis.set_major_formatter(FuncFormatter(thousands))
ax1.set_title('Recaudación Total por Mes', fontsize=20)
ax1.set_ylim(0)

# Gráficos de líneas principales de promedio de ocupación
line1, = ax2.plot(df_2022['Mes'], df_2022['Promedio índice de ocupación'], label='2022', marker='o', alpha=0.3)
line2, = ax2.plot(df_2023['Mes'], df_2023['Promedio índice de ocupación'], label='2023', marker='o', alpha=0.3, color='brown')
line3, = ax2.plot(df_2024['Mes'], df_2024['Promedio índice de ocupación'], label='2024', marker='o', alpha=0.3)
ax2.yaxis.set_major_formatter(FuncFormatter(format_percent))
ax2.set_xlabel('Mes')
ax2.set_title('Promedio índice de ocupación por Mes', fontsize=20)
ax2.set_ylim(0)

# Añadir colores de fondo para los meses de invierno y verano
for idx in indices_invierno:
    ax1.axvspan(idx - 0.5, idx + 0.5, color='#f0e555', alpha=0.3)
    ax2.axvspan(idx - 0.5, idx + 0.5, color='#f0e555', alpha=0.3)

for idx in indices_verano:
    ax1.axvspan(idx - 0.5, idx + 0.5, color='#5b1388', alpha=0.3)
    ax2.axvspan(idx - 0.5, idx + 0.5, color='#5b1388', alpha=0.3)

# Función para calcular la media de un atributo en los índices dados
def calcular_media(dataframes, indices, columna):
    valores = []
    for df in dataframes:
        for i in indices:
            if i < len(df):
                valores.append(df[columna][i])
    return sum(valores) / len(valores)

# Calcular medias y añadir líneas horizontales
rmedia_invierno = calcular_media([df_2022, df_2023, df_2024], indices_invierno, 'Recaudación Total')
rmedia_verano = calcular_media([df_2022, df_2023, df_2024], indices_verano, 'Recaudación Total')
for idx in indices_invierno:
    ax1.hlines(rmedia_invierno, idx - 0.5, idx + 0.5, colors='#f0e555', linestyles='dashed', linewidth=3)
for idx in indices_verano:
    ax1.hlines(rmedia_verano, idx - 0.5, idx + 0.5, colors='#5b1388', linestyles='dashed', linewidth=3)

imedia_invierno = calcular_media([df_2022, df_2023, df_2024], indices_invierno, 'Promedio índice de ocupación')
imedia_verano = calcular_media([df_2022, df_2023, df_2024], indices_verano, 'Promedio índice de ocupación')
for idx in indices_invierno:
    ax2.hlines(imedia_invierno, idx - 0.5, idx + 0.5, colors='#f0e555', linestyles='dashed', linewidth=3)
for idx in indices_verano:
    ax2.hlines(imedia_verano, idx - 0.5, idx + 0.5, colors='#5b1388', linestyles='dashed', linewidth=3)

# Añadir leyendas
handles1, labels1 = ax1.get_legend_handles_labels()
ax1.legend(handles=[line3, line2, line1], labels=['2024', '2023', '2022'], loc='lower right', fontsize=16)

handles2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(handles=[line3, line2, line1], labels=['2024', '2023', '2022'], loc='lower right', fontsize=16)

# Etiquetas adicionales
ax1.text(-2.58, 69830, '70,438 € -', color='#5b1388', fontsize=18, ha='left', fontweight='demi')
ax1.text(-2.58, 74427.2036, '75,028 € -', color='#f4d742', fontsize=18, ha='left', fontweight='demi')
ax2.text(-2.3, 40.527, '40.93% -', color='#f0e555', fontsize=18, ha='left', fontweight='demi')
ax2.text(-2.29, 37.773, '38.14% -', color='#5b1388', fontsize=18, ha='left', fontweight='demi')

plt.tight_layout()
plt.show()
```

## Histórico

```python
df = pd.DataFrame(historico)

# Convertir las columnas al formato adecuado
df['Fecha'] = pd.to_datetime(df['Fecha'], format='%d/%m/%Y')

# Multiplicar por 100 para obtener el porcentaje
df['Índice de ocupación'] *= 100

# Eliminar los 8 primeros registros porque están vacíos
df = df[8:]

# Ajustar el tamaño de la fuente globalmente
plt.rc('font', size=14)          # Tamaño de la fuente principal
plt.rc('axes', labelsize=16)     # Tamaño de las etiquetas de los ejes
plt.rc('xtick', labelsize=14)    # Tamaño de las etiquetas de los ticks del eje x
plt.rc('ytick', labelsize=14)    # Tamaño de las etiquetas de los ticks del eje y
plt.rc('legend', fontsize=14)    # Tamaño de la fuente de la leyenda
plt.rc('figure', titlesize=18)   # Tamaño del título de la figura

# Crear la figura y el eje
fig, ax1 = plt.subplots(figsize=(15, 6))

# Graficar la Recaudación como un diagrama de barras en el primer eje
bars = ax1.bar(df['Fecha'], df['Recaudación'], color='orange', label='Recaudación', width=20)

# Crear el segundo eje que comparte el mismo eje x
ax2 = ax1.twinx()

# Graficar el Índice de Ocupación en el segundo eje (línea)
line = ax2.plot(df['Fecha'], df['Índice de ocupación'], color='b', label='Índice de Ocupación')

# Configurar los límites del eje y para que empiecen en 0
ax1.set_ylim(0, max(df['Recaudación']) * 1.1)
ax2.set_ylim(0, max(df['Índice de ocupación']) * 1.1)

# Configurar las etiquetas de los ejes y el título
ax1.set_xlabel('Año')
ax1.set_ylabel('Recaudación', color='orange', fontweight='bold')
ax2.set_ylabel('Índice de Ocupación', color='b', fontweight='bold')
plt.title('Recaudación y Índice de Ocupación a lo largo del tiempo')

# Mostrar solo los años en el eje x
ax1.xaxis.set_major_locator(plt.MaxNLocator(integer=True))

# Rotar las etiquetas del eje x para mejor visualización
plt.xticks(rotation=45)

# Agregar líneas verticales al final de cada año y su leyenda
for year in range(df['Fecha'].dt.year.min(), df['Fecha'].dt.year.max() + 1):
    ax1.axvline(pd.Timestamp(f'{year}-12-31'), color='black', linestyle='--', alpha=0.8)

# Configurar los ticks del eje x y las etiquetas de año en la mitad de cada intervalo de tiempo
years = range(df['Fecha'].dt.year.min(), df['Fecha'].dt.year.max() + 1)
ticks = [pd.Timestamp(f'{year}-06-30') for year in years]  # Colocar los ticks a mitad de año
tick_labels = [str(year) for year in years]  # Etiquetas de los ticks

ax1.set_xticks(ticks)
ax1.set_xticklabels(tick_labels)

# Formatear los ticks del eje y (ax1) con separadores de millar y el símbolo del euro
formatter_euro = mtick.FuncFormatter(lambda x, pos: '{:,.0f} €'.format(x))
ax1.yaxis.set_major_formatter(formatter_euro)

# Formatear los ticks del eje y (ax2) con el símbolo de porcentaje
formatter_percent = mtick.FuncFormatter(lambda x, pos: '{:.0f}%'.format(x))
ax2.yaxis.set_major_formatter(formatter_percent)

# Crear una entrada en la leyenda para las líneas verticales
legend_line = Line2D([0], [0], color='black', linestyle='--', linewidth=2, label='Separadores de Año')

# Añadir la entrada de la línea vertical a la leyenda
ax1.legend(handles=[legend_line], loc='upper left', shadow=True)

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```
